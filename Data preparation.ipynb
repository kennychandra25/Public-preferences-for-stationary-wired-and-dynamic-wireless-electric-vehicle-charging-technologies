## Mapping and Function

```python
column_mapping = {'1.1': 'EVExp', 
                  '0.1': 'Age', 
                  '9.1': 'Gender',
                  '9.6': 'Income',
                  'A1.3': 'EVBrand',
                  '9.3': 'Ethnic',
                  '9.4': 'Education',
                  '9.5': 'Occupation',
                  'A6.7_2': 'ChargeTimePref'
                 }

age_mapping = {
               'Under 18': 'Under 35', 
               '18-24':'Under 35', 
               '25-34': 'Under 35',
               '35-44': '35-54',
               '45-54': '35-54',
               '55-64': 'Over 54',
               'Over 65': 'Over 54'
              }
```

## Data Preparation

```python
sp_df = df.iloc[2:, np.concatenate(([8], [18], [20], [22], [156], [204], np.arange(211,215), np.arange(180, 204)))]

# rename column
sp_df.rename(columns = column_mapping, 
                    inplace = True)

# redefine age
sp_df['Age'] = sp_df['Age'].map(age_mapping)

sp_df.iloc[:3, :15]
```

```python
scenario_question = sp_df.columns.values[10:]

scenario_question
```

```python
# melt the dataframe to fit Apollo data format

melted_sp_df = sp_df.melt(id_vars=['ResponseId', 'Age', 'Gender', 'Income', 'EVExp', 
                                   'EVBrand', 'ChargeTimePref', 'Ethnic', 'Education', 'Occupation'], 
                          value_vars = scenario_question, 
                          var_name= 'Scenario', value_name='Decision')

melted_sp_df['EVExp'] = melted_sp_df['EVExp'].apply(lambda x: 'No EV Experience' if x == 'I have never owned or leased an EV' else 'Have EV Experience')

melted_sp_df[:25]
```

```python
choice_exp = melted_sp_df.dropna(subset = ['Decision']).reset_index(drop=True)
choice_exp[:25]
choice_exp.info()
choice_exp.shape
```
